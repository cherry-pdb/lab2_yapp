```python
import pandas as pd
```

```python
pubg_stats = pd.read_csv('Pubg_Stats.csv')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
lp = sb.lineplot(pubg_stats, x = 'Headshots', y = 'Kills').set_title('Линейная диаграмма отошения убийств в голову ко всем убийствам\n')
lp_xlabel = plt.xlabel('убийства в голову')
lp_ylabel = plt.ylabel('все убийства')
```

# На данной линейной диаграмме я визуализировал зависимость убийств в голову ко всем убийствам. По оси абсцисс я расположил 
# убийства в голову, а по оси ординат - общее количество убийств. Проанализировав построенную линеную диаграмму, я заметил, 
# что по статистике чем больше количество убийств, тем больше убийств в голову, то есть количество убийств в голову 
# прямопропорционально количеству всех убийств. Однако график нестабильно идёт вверх, а значит, что количество убийств в голову
# также зависит от каких-то ещё параметров (например, скилл игрока).

```python
players = pubg_stats.groupby('Rank').sum('Wins')['Wins']
ranks = pubg_stats['Rank'].unique().tolist()
plt.pie(players.tolist(), labels=ranks, autopct='%1.1f%%')
plt.title('Диаграмма распределения игроков по их рангам')
plt.show()
```

# На данной круговой диаграмме я визуализировал распределение игроков по их рангам. Всего в игре 4 ранга: Silver (красный цвет), 
# Gold (синий цвет), Diamond (оранжевый цвет), Platinum (зелёный цвет). Проанализировав построенный пайчарт, я заметил,
# что больше всего (почти половина) в Pubg игроков с рангом Platinum (44.7%), а меньше всего игроков с рангом Silver (2.4%).  
# То есть почти половина игроков уже достигла максимального звания в игре, а новичков осталось очень мало.

```python
lp2 = sb.lineplot(pubg_stats, x = 'Wins', y = 'Matches_Played').set_title('Диаграмма зависимости побед от сыгранных матчей\n')
lp2_xlabel = plt.xlabel('победы')
lp2_ylabel = plt.ylabel('сыгранные матчи')
```

# На данной линейной диаграмме я визуализировал зависимость побед от в целом сыгранных матчей. По оси абсцисс я расположил 
# победы, а по оси ординат - общее количество сыгранных матчей. Проанализировав построенную линеную диаграмму, я заметил, 
# что по статистике чем больше количество сыграно матчей, тем больше побед, то есть количество побед прямопропорционально 
# количеству всех игр. Но стоит заметить, что в середине графика, на 40 победах и, примерно, 275 играх статистика немного упала, 
# значит один или несколько игроков имеют очень плохой винрейт (отношение побед к сыгранным матчам), следовательно винрейт зависит
# также от второстепенных параметров (например, скилл игрока).
