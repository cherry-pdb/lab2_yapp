```python
import pandas as pd
```

```python
pubg_stats = pd.read_csv('Pubg_Stats.csv')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
lp = sb.lineplot(pubg_stats, x = 'Headshots', y = 'Kills').set_title('        \n')
lp_xlabel = plt.xlabel('  ')
lp_ylabel = plt.ylabel(' ')
```

```python
players = pubg_stats.groupby('Rank').sum('Wins')['Wins']
ranks = pubg_stats['Rank'].unique().tolist()
plt.pie(players.tolist(), labels=ranks, autopct='%1.1f%%')
plt.title('     ')
plt.show()
```
