```python
import numpy as np
import pandas as pd
```

```python
cleaned_ships_data = pd.read_csv('Cleaned_ships_data.csv')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
plt.figure(figsize=(15,7))
gist1 = sb.lineplot(cleaned_ships_data, x = "length", y = "dwt", hue='ship_name').set_title(
    'График зависимости средней маскимальной загрузки кораблей от длины корабля\n')
ylabel_lp = plt.ylabel('средняя максимальная загрузка корабля, кг\n')
xlabel_lp = plt.xlabel('длина корабля, м\n')
```

```python
plt.figure(figsize=(15,7))
gist1 = sb.lineplot(cleaned_ships_data, x = "ship_name", y = cleaned_ships_data['length'] * cleaned_ships_data['width']).set_title(
    'График зависимости средней маскимальной загрузки кораблей от типа корабля\n')
xlbl = plt.xlabel('\nтип корабля')
ylbl = plt.ylabel('средняя максимальная загрузка корабля, кг\n')
```

```python
import matplotlib.pyplot as plt
ships = cleaned_ships_data.groupby("ship_name").sum('gt')['gt']
names = cleaned_ships_data["ship_name"].unique().tolist()
plt.pie(ships.tolist(), labels=names, autopct='%1.1f%%')
plt.title('Распределение количества кораблей по их типу')
plt.show()
```

```python
sb.displot(cleaned_ships_data, x = "built_year", color='Pink')
displot = plt.title('Гистограмма распределения количества кораблей по годам постройки\n')
xlabel_dp = plt.xlabel('год постройки')
ylabel_dp = plt.ylabel('количество')
```

```python
dwt_ranges = np.linspace(cleaned_ships_data['dwt'].min(), cleaned_ships_data['dwt'].max(), num = 21)
plt.figure(figsize=(12,6))
plt.hist2d(cleaned_ships_data['built_year'], cleaned_ships_data['dwt'], bins = [40, dwt_ranges], cmap='Greens')
plt.title('Гистограмма распределения количества кораблей по годам постройки и максимальному весу груза\n')
plt.colorbar(label='\nколичество кораблей')
plt.xlabel('годы постройки')
plt.ylabel('максимальный вес груза, кг\n')
plt.show()
```

```python
import pandas as pd
import seaborn as sb

import warnings 
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")

iris = pd.read_csv('IRIS.csv')
pp = sb.pairplot(iris, hue='species')
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sb.violinplot(x="species", y="sepal_length", data=iris)
plt.ylabel('длина чашелистника, см')
plt.subplot(2, 2, 2)
sb.violinplot(x="species", y="sepal_width", data=iris)
plt.ylabel('ширина чашелистника, см')
plt.subplot(2, 2, 3)
sb.violinplot(x="species", y="petal_length", data=iris)
plt.ylabel('длина лепестка, см')
plt.subplot(2, 2, 4)
sb.violinplot(x="species", y="petal_width", data=iris)
plt.ylabel('ширина лепестка, см')
plt.tight_layout()
plt.show()
```
