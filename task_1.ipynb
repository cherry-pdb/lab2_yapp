```python
import numpy as np
import pandas as pd
```

```python
cleaned_ships_data = pd.read_csv('Cleaned_ships_data.csv')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
gist1 = sb.lineplot(cleaned_ships_data, x = "dwt", y = "length").set_title(
    '        \n')
xlabel_lp = plt.xlabel('\n   , ')
ylabel_lp = plt.ylabel(' , \n')
```

```python
import matplotlib.pyplot as plt
ships = cleaned_ships_data.groupby("ship_name").sum('gt')['gt']
names = cleaned_ships_data["ship_name"].unique().tolist()
plt.pie(ships.tolist(), labels=names, autopct='%1.1f%%')
plt.title('     ')
plt.show()
```

```python
sb.displot(cleaned_ships_data, x = "built_year")
displot = plt.title('      \n')
xlabel_dp = plt.xlabel(' ')
ylabel_dp = plt.ylabel('')
```

```python
dwt_ranges = np.linspace(cleaned_ships_data['dwt'].min(), cleaned_ships_data['dwt'].max(), num = 21)
plt.figure(figsize=(12,6))
plt.hist2d(cleaned_ships_data['built_year'], cleaned_ships_data['dwt'], bins = [40, dwt_ranges], cmap='Blues')
plt.title('          \n')
plt.colorbar(label='\n ')
plt.xlabel(' ')
plt.ylabel('  , \n')
plt.show()
```
